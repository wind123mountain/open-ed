```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer
from peft import PeftModel

base_model_path = "Qwen/Qwen3-8B" 
lora_path = "results/qwen3/sft_8B/e5-bs1-lr0.0001-G16-N2-NN1-lora-16-64-0.1/294" 
save_path = "merged_model"

# Load base model
base_model = AutoModelForCausalLM.from_pretrained(
    base_model_path,
    torch_dtype=torch.bfloat16,   # hoặc float16
    device_map="cpu"
)

# Load LoRA vào base model
model = PeftModel.from_pretrained(base_model, lora_path)

# Merge LoRA weights vào base weights
model = model.merge_and_unload()

# Save model hoàn chỉnh
model.save_pretrained(save_path)

# Save tokenizer (quan trọng)
tokenizer = AutoTokenizer.from_pretrained(base_model_path)
tokenizer.save_pretrained(save_path)

print("✅ Merge hoàn tất và đã lưu model.")
```
