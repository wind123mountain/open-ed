```python
import ast

def process_log(file_path, label_mapping, streams):
    
    id_to_label = {v: k.lower() for k, v in label_mapping.items()}

    with open(file_path, 'r', encoding='utf-8') as f:
        lines = f.readlines()
        last_line = lines[-1].strip()

    dict_str = last_line.split("|", 2)[2].strip()
    log_data = ast.literal_eval(dict_str)
    
    trigger_metrics = log_data.get('trigger_per_type', {})

    print(f"{'TASK':<10} | {'SỐ LABEL':<10} | {'AVG F1 SCORE':<15}")
    print("-" * 40)

    for task_idx, task_ids in enumerate(streams):
        f1_scores = []
        valid_labels = []

        for label_id in task_ids:
            if label_id == 0:  # Bỏ qua ID 0
                continue
            
            label_name = id_to_label.get(label_id)
            if label_name:
                valid_labels.append(label_name)
                if label_name in trigger_metrics:
                    f1_scores.append(trigger_metrics[label_name]['f1'])

        avg_f1 = sum(f1_scores) / len(f1_scores) if f1_scores else 0.0
        print(f"Task {task_idx + 1:<5} | {len(valid_labels):<10} | {avg_f1:.4f}")
```

```python
import json

with open("data/ext-data/ace/label2id.json", "r", encoding="utf-8") as f:
    label2id = json.load(f)

with open("data/ext-data/ace/streams.json", "r", encoding="utf-8") as f:
    streams = json.load(f)


process_log("results/qwen3/v3/ace_0_0.6B_cl/4/log.txt", label2id, streams)
```
