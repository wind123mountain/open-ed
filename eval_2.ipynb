```python
from sentence_transformers import SentenceTransformer
import torch

# Download from the ðŸ¤— Hub
model = SentenceTransformer("google/embeddinggemma-300m")

event_type_descriptions_embed = model.encode([f"{k}: {v}" for k, v in event_type_descs.items()])
```

```python
import json


gt_trigger = []
with open("processed_data/ace/qwen/test.jsonl", "r") as f:
    for line in f:
        events = json.loads(json.loads(line)['response'])['events']
        e_gt = []
        triggers = []
        for trigger in events:
            triggers.append((trigger['trigger_text'], trigger['type']))

        gt_trigger.append(triggers)
```

```python
all_pred_texts = []   # Chá»©a toÃ n bá»™ text Ä‘á»ƒ Ä‘áº©y vÃ o GPU 1 lÆ°á»£t
metadata = []         # Chá»©a thÃ´ng tin gá»‘c (t_text, t_type, t_desc) Ä‘á»ƒ map láº¡i sau
trigger_counts = []   # LÆ°u sá»‘ lÆ°á»£ng trigger cá»§a má»—i dÃ²ng JSON Ä‘á»ƒ tÃ¡i táº¡o cáº¥u trÃºc ban Ä‘áº§u

with open("eval_outputs/results/qwen2.5/distillm_0.5B_4B_on_srkl/490/ace_answers.jsonl", "r") as f:
    for line in f:
        try:
            events = json.loads(json.loads(line)['text'])['events']
            trigger_counts.append(len(events)) # Ghi nhá»› dÃ²ng nÃ y cÃ³ bao nhiÃªu trigger
            
            for trigger in events:
                t_text = trigger.get('trigger_text', '')
                t_type = trigger.get('type', '')
                t_desc = trigger.get("description", "")
                
                all_pred_texts.append(f"{t_type}: {t_desc}")
                metadata.append((t_text, t_type, t_desc))
                
        except Exception:
            trigger_counts.append(0) # DÃ²ng bá»‹ lá»—i/khÃ´ng cÃ³ data thÃ¬ ghi nháº­n lÃ  0 trigger

# 3. BÆ°á»›c Xá»­ lÃ½ song song trÃªn GPU (Batch Encoding)
pred_trigger = []

if all_pred_texts:
        
    pred_embeds = model.encode(
        all_pred_texts, 
        batch_size=128,  
        show_progress_bar=True # Hiá»‡n thanh tiáº¿n trÃ¬nh cho dá»… theo dÃµi
    )
    
    # 4. TÃ­nh Cosine Similarity báº±ng phÃ©p nhÃ¢n ma tráº­n (Cá»±c nhanh)
    # pred_embeds shape: (tá»•ng_sá»‘_triggers, hidden_size)
    # ace_embeds shape: (sá»‘_lÆ°á»£ng_ace_types, hidden_size)
    # cosine_scores shape: (tá»•ng_sá»‘_triggers, sá»‘_lÆ°á»£ng_ace_types)
    similarity_scores = model.similarity(pred_embeds, event_type_descriptions_embed)
    
    # Láº¥y max score vÃ  index cho Táº¤T Cáº¢ cÃ¡c triggers cÃ¹ng lÃºc
    max_scores_tensor, max_indices_tensor = torch.max(similarity_scores, dim=1)
    
    # Chuyá»ƒn vá» list cá»§a Python Ä‘á»ƒ duyá»‡t cho nhanh
    max_scores = max_scores_tensor.tolist()
    max_indices = max_indices_tensor.tolist()

    # 5. TÃ¡i táº¡o láº¡i cáº¥u trÃºc máº£ng 2 chiá»u ban Ä‘áº§u
    flat_index = 0
    for i, count in enumerate(trigger_counts):
        current_triggers = []
        # Láº¥y Ä‘Ãºng sá»‘ lÆ°á»£ng trigger thuá»™c vá» dÃ²ng JSON nÃ y
        for _ in range(count):
            t_text, t_type, t_desc = metadata[flat_index]
            score = round(max_scores[flat_index], 4)
            best_idx = max_indices[flat_index]
            
            current_triggers.append((t_text, t_type, event_keys[best_idx], score))
            flat_index += 1
            if t_type != event_keys[best_idx]:
                print(i, t_text, t_type, event_keys[best_idx], gt_trigger[i])
            
        pred_trigger.append(current_triggers)

else:
    # Xá»­ lÃ½ trÆ°á»ng há»£p file rá»—ng
    pred_trigger = [[] for _ in trigger_counts]
```
